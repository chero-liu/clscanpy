```r
library(dplyr)
source('/gpfs/oe-scrna/liuchenglong/RaD/pyscenic/netplot/tools/utils.r')
source("~/script/lclFunc.r")

input = '/gpfs/oe-scrna/liuchenglong/RaD/pyscenic/test/DZOE2024091605/pySCENIC/data/expr_mat.adjacencies.tsv'
interestTF = 'E2f7,E2f1'
outdir = '/gpfs/oe-scrna/liuchenglong/RaD/pyscenic/netplot/test'
targetTop = 100
interestTFColor = '#1E90FF'
interestTFSize = 10
showTargetTop = 0.3
topTargetColor = '#E87D72'
topTargetSize = 3
highlightTarget = 'Syce2,Donson,Pdxp'
highlightTargetColor = '#00aa55'
highlightTargetSize = 6


metadata = read.csv(input,sep = '\t')
interestTF = standardizeDelimitedVector(interestTF)
highlightTarget = standardizeDelimitedVector(highlightTarget)
targetTop = as.numeric(targetTop)
createDir(outdir)

edge_df = process_tf(interestTF,metadata,targetTop,highlightTarget)


edge_df_top = edge_df %>%
  group_by(TF) %>%
  slice_max(order_by = importance, n = targetTop*showTargetTop,
           with_ties = FALSE) %>%
  ungroup()
showTarget = unique(edge_df_top$target)


showLabel = unique(c(interestTF,showTarget,highlightTarget))
allGenes = unique(c(unique(edge_df$target),unique(edge_df$TF)))
vertex_df = data.frame(name=allGenes,label='')

vertex_df[vertex_df$name %in% showLabel,'label'] = vertex_df[vertex_df$name %in% showLabel,'name']

vertex_df$size = ifelse(vertex_df$name %in% interestTF,interestTFSize,ifelse(vertex_df$name %in% showTarget,topTargetSize,ifelse(vertex_df$name %in% highlightTarget,highlightTargetSize,1)))
vertex_df$color = ifelse(vertex_df$name %in% interestTF,interestTFColor,ifelse(vertex_df$name %in% showTarget,topTargetColor,ifelse(vertex_df$name %in% highlightTarget,highlightTargetColor,"#FFFFFF")))

if(highlightTarget != 'no'){
    numb = 4
}else{
    numb = 3
}

pdf(paste0(outdir,'/TFsNetRegulation.pdf'),width = 12,height = 8)
net_plot(edge_df,
vertex_df,
layout = 'layout_with_lgl',
edge.color = '#00CED1',
vertex.label.cex=0.8,
vertex.size = vertex_df$size,
vertex.color = vertex_df$color,
legend.vertex.label = c('TF',paste0('Top ',showTargetTop*100,'% of target'),'Other target','highlightTarget')[1:numb],
legend.vertex.color = c(interestTFColor,topTargetColor,"#FFFFFF",highlightTargetColor)[1:numb])
dev.off()


png(paste0(outdir,'/TFsNetRegulation.png'),width = 1200,height = 800)
net_plot(edge_df,
vertex_df,
layout = 'layout_with_lgl',
edge.color = '#00CED1',
vertex.label.cex=0.8,
vertex.size = vertex_df$size,
vertex.color = vertex_df$color,
legend.vertex.label = c('TF',paste0('Top ',showTargetTop*100,'% of target'),'Other target','highlightTarget')[1:numb],
legend.vertex.color = c(interestTFColor,topTargetColor,"#FFFFFF",highlightTargetColor)[1:numb])
dev.off()
```
